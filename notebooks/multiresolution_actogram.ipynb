```python
import os,glob,json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime,timedelta
```

```python
def load_json(root):
    with open(root) as file:
        detections= json.load(file)
    return detections
```

```python
def check_label(t,trk_class):
    
    label=[]
    if t in trk_class['pollen']: label.append('pollen')
    if t in trk_class['foraging']['leaving']:label.append('leaving')
    if t in trk_class['foraging']['entering']:label.append('leaving')
    if t in trk_class['foraging']['entering-leaving']:label.append('entering-leaving')
    if t in trk_class['foraging']['walking']:label.append('walking')
    for k in trk_class['track_shape']:
        if t in track_class['track_shape'][k]:
            label.append(k)
    return label 
```
